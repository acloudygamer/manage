在使用markdown的过程中，我们需要了解一定的markdown格式才能更好的在GitHub上进行项目管理
1. 标题
使用 # 表示标题，# 的数量表示标题的级别（1-6级）
2. 段落和换行
段落之间需要空一行。

如果需要换行，可以在行尾添加两个空格。
3. 加粗和斜体
用 ** 或 __ 表示加粗。

用 * 或 _ 表示斜体。
4. 列表
无序列表
使用 -、* 或 + 表示无序列表。
有序列表
使用数字加点表示有序列表。
5. 链接和图片
链接：[显示文本](链接地址)

图片：![替代文本](图片地址)
6. 引用
使用 > 表示引用。
7. 代码
单行代码：用反引号 ` 包裹。

多行代码：用三个反引号 包裹。
8. 分割线
使用 --- 或 *** 表示分割线。
