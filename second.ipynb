## 在使用markdown的过程中，我们需要了解一定的markdown格式才能更好的在GitHub上进行项目管理
1. 标题
使用 # 表示标题，# 的数量表示标题的级别（1-6级）
2. 段落和换行
    1. 段落之间需要空一行。  
    2. 如果需要换行，可以在行尾添加两个空格。
3. 加粗和斜体
    1. 用 ** 或 __ 表示加粗。  
    2. 用 * 或 _ 表示斜体。
4. 列表
    1. 无序列表     
   使用 -、* 或 + 表示无序列表。     
    2. 有序列表  
    使用数字加点表示有序列表。
5. 链接和图片
链接：[显示文本](链接地址)  
图片：![替代文本](图片地址)
1. 引用
使用 > 表示引用。
1. 代码
单行代码：用反引号 ` 包裹。  
多行代码：用三个反引号 包裹。
1. 分割线
使用 --- 或 *** 表示分割线。
